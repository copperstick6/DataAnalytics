(7) Let's explore the cyclicality of avocado prices over the year. Create a new variable that contains the mean price of an avocado in that week across all years of the data for organic avocados and for conventional avocados. (To help see the relationship between week and date use the head function to print rows of the data set ex: head(organics, n=100)) Once you have created the new meaned price variable, plot the meaned price on the Y-axis and week on the x-axis with a fitted curve for both organic and conventional. Do organic and conventional avocados have similar price cyclicality over the course of a year?

```r
variables = c("week", "averageprice")
data_conventional_means = data_conventional[variables]
data_conventional_means<-aggregate(data_conventional_means, by = list(data_conventional_means$week), FUN = mean)
head(data_conventional_means)

data_organic_means = data_organic[variables]
data_organic_means<-aggregate(data_organic_means, by = list(data_organic_means$week), FUN = mean)
head(data_organic_means)
```

```r
P <- ggplot(data_conventional_means,aes(week, averageprice)) + geom_point() + xlab("Week") + ylab("Mean Average Price")
P + ggtitle("Scatterplot and fitted curve of week vs mean average price for conventional avocadoes") + geom_smooth(method='lm')

P <- ggplot(data_organic_means,aes(week, averageprice)) + geom_point() + xlab("Week") + ylab("Mean Average Price")
P +  ggtitle("Scatterplot and fitted curve of week vs mean average price for organic avocadoes") + geom_smooth(method='lm')
```

(8) Now let's look at the relationship between the price of avocados and the quantity of avocados sold. Regress the total volume of avocados sold on the average price of avocados for both organic and conventional avocados. 

```r
model1 <- lm(data_conventional$totalvolume ~ data_conventional$averageprice)
summary(model1)

model2 <- lm(data_organic$totalvolume ~ data_organic$averageprice)
summary(model2)
```
